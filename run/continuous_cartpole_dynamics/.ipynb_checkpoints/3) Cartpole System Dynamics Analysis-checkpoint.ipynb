```python
import pickle
import numpy as np
```

```python
with open('./cartpole_system_model/best_controller.pkl', 'rb') as filepath:
    save_best_controller_dict = pickle.load(filepath)
```

```python
T = save_best_controller_dict["Target"]
print(T)
```

```python
w, V = np.linalg.eig(T)
for i in range(len(w)):
    print("Eigenvalue: ", i)
    print(w[i])
    print("Eigenvector: ", i)
    print(V[:, i])
    print("\n")
```

### Cart Position and Velocity

The eigenvalues for cart position and cart velocity are real.

Both are negative, suggesting a stable fixed point at the origin.

The cart velocity has a higher convergence rate than the velocity.

### Pole Angle and Angular Velocity

The eigenvalues for the pole angle and pole angular velocity are real.

Both are negative, suggesting a stable fixed point at the origin.

The pole angular velocity has a higher convergence rate than the angle.

```python
import matplotlib.pyplot as plt
```

```python
x_pos = [V[0, 0] * -10, V[0, 0] * 10]
y_pos = [V[1, 0] * -10, V[1, 0] * 10]
plt.plot(x_pos, y_pos, label = "Cart eigenvector")

x_pos_vel = [V[0, 1] * -10, V[0, 1] * 10]
y_pos_vel = [V[1, 1] * -10, V[1, 1] * 10]
plt.plot(x_pos_vel, y_pos_vel, label = "Cart velocity eigenvector")

plt.plot([-8, 8], [0, 0], color = "black")
plt.plot([0, 0], [-8, 8], color = "black")

plt.xlabel(r"$x$")
plt.ylabel(r"$\dot x $")

plt.legend()
plt.show()
```

```python
x_ang = [V[2, 2] * -10, V[2, 2] * 10]
y_ang = [V[3, 2] * -10, V[3, 2] * 10]
plt.plot(x_ang, y_ang, label = "Angle eigenvector")

x_ang_vel = [V[2, 3] * -10, V[2, 3] * 10]
y_ang_vel = [V[3, 3] * -10, V[3, 3] * 10]
plt.plot(x_ang_vel, y_ang_vel, label = "Angular velocity eigenvector")

plt.plot([-8, 8], [0, 0], color = "black")
plt.plot([0, 0], [-8, 8], color = "black")

plt.xlabel(r"$\theta$")
plt.ylabel(r"$\dot \theta $")

plt.legend()
plt.show()
```
